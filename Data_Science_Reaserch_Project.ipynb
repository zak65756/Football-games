Question: Do NFL teams score more when playing at home games?

```python
# Total score for each game
df['total_score'] = df['home_score'] + df['away_score']
df['home_win'] = df['home_score'] > df['away_score']
```

```python
# Calculate average points scored at home
home_avg = df['home_score'].mean()

# Calculate average points scored away
away_avg = df['away_score'].mean()

# Print the results
print("Average Home Score:", round(home_avg, 2))
print("Average Away Score:", round(away_avg, 2))
```
